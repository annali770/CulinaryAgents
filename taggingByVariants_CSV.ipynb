```python
import pandas as pd
```

```python
#reads CSV files using pandas to allow viewing/editing
d = pd.read_csv("combinedDescriptions.csv")
v = pd.read_csv("variants.csv")
```

```python
#drops column by index if needed, inplace saves it for further use (remove if just to view data and not edit)
#d.drop(d.columns[0], axis=1, inplace=True)

#drops row by index
#d.drop([2], inplace=True)

#replaces entire column with uniform value (ig. empty space), creates new column if not existing
#d['skills'] = ""

#saves edits to CSV -- important!
v.to_csv('variants.csv')
```

```python
#creates variable to keep track of row indices in csv to be edited
dIterate = 0

#loops thorough each value of specific column in targeted CSV
for desc in d['description']:
    
    #breaks after max number of rows in CSV so out-of-bounds error doesn't occur
    if dIterate == 13560:
        break
        
    #creates variable to keep track of row indices in csv to be read
    vIterate=0
    
    #loops through each value of specific column in second CSV
    for var in v['variant']:
        
        #determines if string in second CSV is within the string in first CSV; if true, returns index of found value, -1 if false
        if str(desc).find(str(var))>-1:
            
            #accounts for potential repeats in tagging if multiple variants of same skill found
            if str(d.iat[dIterate,7]).find(str(v.iat[vIterate,1])) <0:
                
                #edits the value in newly created column with each new skill, "||" as seperator
                #function df.iat[row, column] allows for editing of original CSV; regular assignments like df[col][row] = value are shallow
                d.iat[dIterate,7] = str(d.iat[dIterate,7]) + "||" + str(v.iat[vIterate,1])
        
        vIterate = vIterate +1
    dIterate = dIterate+1
```
